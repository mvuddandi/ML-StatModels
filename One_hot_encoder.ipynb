```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df.info()
```

```python
town_dummy = pd.get_dummies(df.town, dtype=int)
town_dummy
```

```python
df_merged = pd.concat([df,town_dummy],axis="columns")
df_merged
```

```python
df_mer1 = df_merged
```

```python
df_mer1 = df_mer1.drop(['town','west windsor'], axis='columns')
```

```python
df_mer1
```

```python
model = linear_model.LinearRegression()
```

```python
X = df_mer1.drop('price', axis='columns')
y = df_mer1.price
model.fit (X,y)
```

```python
model.predict([[3000,1,0]])
```

```python
model.score (X,y)
```

Using the one hot encode



```python
from sklearn .preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
df_mer2 = df
df_mer2
```

```python
df_mer2.town = le.fit_transform(df_mer2.town)

df_mer2
```

```python
X = df_mer2[['town','area']].values
X
```

```python
y = df_mer2.price
y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse_output=False), ['town'])
    ],
    remainder='passthrough'
)

X_transformed = ct.fit_transform(df)
```

```python
X_transformed
```

```python
X_transformed = X_transformed[:,1:]
X_transformed
```

```python
model.fit(X_transformed,y)
```

```python
model.predict([[1,0,2600]])
```
